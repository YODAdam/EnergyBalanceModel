```python
import numpy as np
import matplotlib.pyplot as plt


Cp = 1.0e10
dt = 1.0e8

S = 1361
alpha = 0.30
sigma = 5.67e-8
epsilon = 0.61

# Initial condition
Tn = 273.0    
nt = 200

time = np.zeros(nt)
temperature = np.zeros(nt)

# Time loop
for n in range(nt):

    time[n] = n * dt / (365 * 24 * 3600)  # convert to years

    E_in = S * (1 - alpha)
    E_out = 4 * sigma * epsilon * Tn**4

    Tend = (dt / Cp) * (E_in - E_out)
    Tn = Tn + Tend

    temperature[n] = Tn

# Plot
plt.figure()
plt.plot(time, temperature - 273.15)
plt.xlabel("Time (years)")
plt.ylabel("Temperature (Â°C)")
plt.title("EBM2 Temperature Evolution")
plt.show()
```
